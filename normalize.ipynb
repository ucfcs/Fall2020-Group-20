```python
#Importing data from csv
from pandas import read_csv
import pandas as pd

terran_data = read_csv('./Output/TerranStat.csv', header=0, index_col=False)

#Dropping the MatchID
terran_data = terran_data.drop(terran_data.columns[0], axis=1)
```

```python
#Tried to create a scatter plot matrix but we have too many variables :(
import matplotlib.pyplot as plt

pd.plotting.scatter_matrix(terran_data.loc[:,"Second":"VespeneLost"])
```

```python
#Some visualization of our data set

terran_data.loc[:,"Second":"VespeneLost"].plot()
plt.show()
```

```python
#Normalization

from sklearn.preprocessing import MinMaxScaler

no_normalize = ['Result','Player','Race','Opponent','Map']

#Names of Columns that we want to Normalize (Numerical variables only)
normalize_columns = [x for x in list(terran_data) if x not in no_normalize]

#Normalizing these columns
norm = terran_data[normalize_columns].values
norm_scaled = MinMaxScaler().fit_transform(norm)
terran_data_normalized = pd.DataFrame(norm_scaled, columns=normalize_columns, index=terran_data.index)

#One-hot Encoding the categorical variables
terran_data_normalized = terran_data_normalized.join(pd.get_dummies(terran_data['Map'], prefix='Map'))
terran_data_normalized = terran_data_normalized.join(pd.get_dummies(terran_data['Result'], prefix='Result'))
terran_data_normalized = terran_data_normalized.join(pd.get_dummies(terran_data['Opponent'], prefix='Opponent'))
terran_data_normalized = terran_data_normalized.join(pd.get_dummies(terran_data['Race'], prefix='Race'))
```

```python
#Printing the Normalized Columns
display(terran_data_normalized)
```

```python
#Standardization

from sklearn.preprocessing import StandardScaler

no_standardize = ['Result','Player','Race','Opponent','Map']

#Names of Columns that we do want to Standardize
standardize_columns = [x for x in list(terran_data) if x not in no_normalize]

#Standardizing these columns
stand = terran_data[standardize_columns].values
stand_scaled = StandardScaler().fit_transform(stand)
terran_data_standardized = pd.DataFrame(stand_scaled, columns=standardize_columns, index=terran_data.index)

#One-hot Encoding the categorical variables
terran_data_standardized = terran_data_standardized.join(pd.get_dummies(terran_data['Map'], prefix='Map'))
terran_data_standardized = terran_data_standardized.join(pd.get_dummies(terran_data['Result'], prefix='Result'))
terran_data_standardized = terran_data_standardized.join(pd.get_dummies(terran_data['Opponent'], prefix='Opponent'))
terran_data_standardized = terran_data_standardized.join(pd.get_dummies(terran_data['Race'], prefix='Race'))
```

```python
#Printing the Standardized Columns
display(terran_data_standardized)
```

```python
#Printing the Standardized mean and std, should be mean=0 and std=1

print('Standardized Mean: ', stand_scaled.mean(axis=0))
print('Standardized Standard Deviation: ', stand_scaled.std(axis=0))
```
